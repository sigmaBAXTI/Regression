5-qadam: Vizualizatsiya

```python
ustun_tanlash = ['X1', 'X2', 'X3', 'X4', 'X5']
selected_df = df[ustun_tanlash]

correlation_matrix = selected_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=10)
plt.title('Korrelyatsiya Matritsasi')
plt.show()
```

```python
sns.scatterplot(x='X1', y='X2', data=selected_df)
plt.show()
```
